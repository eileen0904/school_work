```python
import requests
from pandas import DataFrame
import pandas as pd
import re
from bs4 import BeautifulSoup

url = 'https://sys.ndhu.edu.tw/SA/XSL_ApplyRWD/ActApply.aspx?lang=en'

html = requests.get(url)

bs = BeautifulSoup(html.content, "html.parser")

names = []
hours = []
limits = []
registers = []
avtivitys = []

for i in range(0, 10):
    name = str(bs.select_one(
        '#BodyContent_gvActs_lblGv_act_name_' + str(i)).string)
    hour = str(bs.select_one(
        '#BodyContent_gvActs_lblGv_xsl_check_' + str(i)).string)
    limit = str(bs.select_one(
        '#BodyContent_gvActs_lblGv_reg_num_' + str(i)).string)
    register = str(bs.select_one(
        '#BodyContent_gvActs_lblGv_reg_dt_' + str(i)).string)
    avtivity = str(bs.select_one(
        '#BodyContent_gvActs_lblGv_act_dt_' + str(i)).string)

    hour = hour.strip('Can')
    hour = hour.replace('/', '')

    limit = limit.replace('People', '')
    limit = re.search(r"\d+", limit).group()

    names.append(name)
    hours.append(hour)
    limits.append(limit)
    registers.append(register)
    avtivitys.append(avtivity)

data = {'Name of the activity': names, 'Certificated': hours, 'Number of people who have registered': limits,

        'Date and time of registeration': registers, 'Date and time of activity': avtivitys}

df = pd.DataFrame(data, index=range(1, 11))
df
```
