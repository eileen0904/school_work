```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/home/xin0/Downloads/data.csv")

town, households, male, female, total, immigration, emigration, birth, death = [], [], [], [], [], [], [], [], []

for i in range(len(df)): #len = row num = 18
    town.append(df["Township"][i])
    households.append(df["Number_of_Households"][i])
    male.append(df["Male"][i])
    female.append(df["Female"][i])
    total.append(df["Total(Male+Female)"][i])
    immigration.append(df["Immigration"][i])
    emigration.append(df["Emigration"][i])
    birth.append(df["Birth"][i])
    death.append(df["Death"][i])

plt.figure(figsize = (13, 13))

plt.subplot(221) #nrows, ncols, index
plt.plot(town, male, label = "Male")
plt.plot(town, female, '-.', label = "Female")
plt.plot(town, total, '--', label = "Total")
plt.legend(loc = (1)) #upper right
plt.title("Population by Gender and Township")
plt.xlabel("Township")
plt.ylabel("Population")
plt.xticks(rotation = "vertical") #rotation = 90

plt.subplot(222)
plt.plot(town, immigration, label = "Immigration")
plt.plot(town, emigration, '--', label = "Emigration")
plt.legend(loc = (1))
plt.title("Immigration and Emigration by Township")
plt.xlabel("Township")
plt.ylabel("Number of People")
plt.xticks(rotation = "vertical")

plt.subplot(223)
plt.bar(town, households)
plt.title("Household Number by Township")
plt.xlabel("Township")
plt.ylabel("Number of Number_of_Households")
plt.xticks(rotation = "vertical")

plt.subplot(224)
plt.plot(town, birth, label = "Birth")
plt.plot(town, death, '--', label = "Death")
plt.legend(loc = (1))
plt.title("Birth and Death by Township")
plt.xlabel("Township")
plt.ylabel("Number of People")
plt.xticks(rotation = "vertical")

plt.tight_layout()
plt.show()
```
